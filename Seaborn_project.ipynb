# **Practice Set on seaborn**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sales = pd.read_csv("/content/Sales Dataset.csv")
sales
```

```python
sns.histplot(data=sales,x='Total Amount',y='Quantity')
plt.grid(True)
```

```python
sns.barplot(data=sales,x='Total Amount',y='Quantity',hue='Product Category')
plt.grid(True)
```

```python
sns.lineplot(data=sales,x='Total Amount',y='Quantity',hue='Product Category', color= 3)
plt.grid(True)
```

```python
numeric_cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
plt.figure(figsize=(6, 4))
sns.heatmap(sales[numeric_cols].corr(), annot=True, cmap="Blues", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
if 'Product Category' in sales.columns and 'Total Amount' in sales.columns:
    plt.figure()
    sns.barplot(data=sales, x='Product Category', y='Total Amount', hue='Gender', errorbar=None, palette='muted')
    plt.title("Average Revenue per Product Category split by Gender")
    plt.ylabel("Average Total Amount")
    plt.show()
```

```python
if 'Total Amount' in sales.columns:
    plt.figure()
    sns.histplot(data=sales, x='Total Amount', kde=True, color='purple', bins=30)
    plt.title("Distribution of Total Purchase Amounts")
    plt.show()
```
